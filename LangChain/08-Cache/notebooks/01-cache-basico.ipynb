# Tutorial 01: Cache Básico - InMemory e SQLite

**Objetivos de Aprendizagem:**
- Entender o que é cache e por que usar
- Implementar cache em memória (InMemory)
- Implementar cache persistente (SQLite)
- Comparar performance e economia de custos

**Pré-requisitos:**
- Tutorial 01-LLMs-Basicos: Introdução aos LLMs concluído
- Compreensão básica de Python

---

## Introdução

O cache é uma funcionalidade essencial do LangChain que permite armazenar respostas de LLMs para reduzir custos e melhorar performance.
