# Tutorial: Reconhecimento Facial e Análise de Expressões Emocionais em Vídeos

Este tutorial prático ensina passo a passo como implementar reconhecimento facial e análise de expressões emocionais em vídeos usando Python, OpenCV, DeepFace e face_recognition.

## Objetivos
- Aprender a detectar faces em vídeos
- Analisar expressões emocionais usando DeepFace
- Reconhecer pessoas conhecidas em vídeos
- Combinar detecção, reconhecimento e análise emocional

## Pré-requisitos
- Python 3.8+
- Bibliotecas: opencv-python, deepface, face-recognition, numpy, tqdm
